9. Identify the offices (by city) that have the least employees and the most employees (answer should include city, numEmps)

```python
con = mysql.connector.connect(user='hbs-student', password='super-secure-password?', host='hbs-ds-060120.c1doesqrid0e.us-east-1.rds.amazonaws.com', database='classicmodels')
df1 = pd.read_sql_query('''SELECT city, count(*) AS numEmps
                           FROM employees
                           JOIN offices USING (officeCode)
                           GROUP BY officeCode, city
                           HAVING numEmps IN (
                               SELECT max(numEmps)
                               FROM (SELECT officeCode, COUNT(*) as numEmps
                                     FROM employees
                                     GROUP BY officeCode) AS max_count
                                     
                               UNION
                                     
                               SELECT min(numEmps)
                               FROM (SELECT officeCode, count(*) as numEmps
                                     FROM employees
                                     GROUP BY officeCode) AS min_count
                                     )
''', con)
df1.head()
```

12. List the product number, product name, and customer name such that the customer has never ordered that product. Order by customerName, productCode.

```python
con = mysql.connector.connect(user='hbs-student', password='super-secure-password?', host='hbs-ds-060120.c1doesqrid0e.us-east-1.rds.amazonaws.com', database='classicmodels')
df1 = pd.read_sql_query('''SELECT *
                           FROM (SELECT p.productCode, p.productName, c.customerNumber, c.customerName
                                 FROM products p
                                 CROSS JOIN customers c) all_combos
                           LEFT JOIN(SELECT p.productCode, o.customerNumber
                                     FROM products p
                                     JOIN orderdetails od USING (productCode)
                                     JOIN orders o USING (orderNumber)
                                     ) good_combos
                           ON all_combos.productCode = good_combos.productCode
                           AND all_combos.customerNumber = good_combos.customerNumber
                           WHERE good_combos.productCode IS NULL
                           ORDER BY customerName, all_combos.productCode
''', con)
df1.head()
```
